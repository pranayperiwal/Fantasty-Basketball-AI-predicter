```python
""" Getting the Datasets into Dataframes """

import glob
import pandas as pd

#dfs with all csv files in pandas
dfs = [pd.read_csv("simulation.csv")] 
simulation_df = dfs[0]
display(simulation_df)
```

```python
""" Removing the Second Position """

simulation_df['POS'] = simulation_df['POS'].apply(lambda x:str(x)[0])
display(simulation_df)
```

```python
""" One Hot Encoding for Start Position """
from sklearn.preprocessing import OneHotEncoder
simulation_df['POS']= simulation_df['POS'].fillna('NO')


encoded_features = []

#one hot encoding the start position for each player
encoded_feat = OneHotEncoder().fit_transform(simulation_df['POS'].values.reshape(-1, 1)).toarray()
n = simulation_df['POS'].nunique()
cols = ['{}_{}'.format('START_POSITION', n) for n in range(1, n + 1)]
encoded_df = pd.DataFrame(encoded_feat, columns=cols)
encoded_df.index = simulation_df.index
encoded_features.append(encoded_df)

#adding encoded features to df
simulation_df = pd.concat([simulation_df, *encoded_features[:1]], axis=1)

#removing start posiiton from df as it has been hot encoded
#simulation_df.drop(['POS'],axis='columns',inplace=True)
display(simulation_df)
```

```python
""" Creating START_POSITION_4 Column """

simulation_df['START_POSITION_4']=0.0
display(simulation_df)
```

```python
""" Only Taking Columns that are Necessary """
simulation_df = simulation_df[['FULL NAME','TEAM','POS','FG_PCT','FG3_PCT','FT_PCT','REB','AST','STL','BLK','START_POSITION_1','START_POSITION_2','START_POSITION_3','START_POSITION_4','PTS/SEC']]
display(simulation_df)
```

```python
""" Saving the New CSV File """

simulation_df.to_csv('/content/drive/MyDrive/COMP3359/AI_Project/data/finalSimulation.csv', index = False)
```
